```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.models import Sequential

train=pd.read_csv("engtamilTrain.csv")
train=train.drop(["Unnamed: 0"],axis=1)
english_sentences=train["en"]
tamil_sentence=train['ta']
english_sentences=english_sentences.head(1000)
tamil_sentences=tamil_sentence.head(1000)
```

```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot
```

```python
def sentToken(sentence):
        # define your dataset
    dataset = sentence

    # convert each sentence to a list of words
    sentences = [sentence.split() for sentence in dataset]

    # print the result
    print(sentences)
    
    return sentences
```

```python
engSentence=sentToken(english_sentences)
```

```python
tamSentence=sentToken(tamil_sentences)
```

```python
def ownWordModel(langsentence,modelname):
    from gensim.models import Word2Vec
    from sklearn.decomposition import PCA
    import matplotlib.pyplot as plt
    model = Word2Vec(langsentence, min_count=1)
    print(model)
    # save model
    model.save(modelname)
    # fit a 2d PCA model to the vectors
    X = model.wv.vectors  # corrected line
    pca = PCA(n_components=2)
    result = pca.fit_transform(X)

    # create a scatter plot of the projection
    plt.scatter(result[:, 0], result[:, 1])
    words = list(model.wv.index_to_key)
    for i, word in enumerate(words):
        plt.annotate(word, xy=(result[i, 0], result[i, 1]))
    plt.show()
```

```python
ownWordModel(engSentence,"engmodel.bin")
```

```python
ownWordModel(tamSentence,"tammodel.bin")
```

#### Using PCA (Principal Component Analysis) to fit a 2D model to word vectors for visualization serves several important purposes in the context of natural language processing (NLP) and machine learning. 

#### Word embeddings are typically high-dimensional vectors (e.g., 100 to 300 dimensions). Visualizing such high-dimensional data directly is not feasible. PCA reduces the dimensionality of the data to 2 or 3 dimensions, making it possible to visualize and interpret the relationships between words in a more comprehensible way

#### fitting a 2D PCA model to word vectors for visualization is a powerful technique that enhances our ability to interpret and analyze word embeddings. It provides insights into the relationships between words, helps evaluate the quality of the embeddings, and facilitates further analysis and machine learning tasks.
